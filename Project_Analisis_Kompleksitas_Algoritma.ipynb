### *Apa solusi yang dapat diterapkan untuk mengatasi masalah pada sistem pembelian tiket tersebut?*

Pembelian tiket konser dengan permintaan tinggi, seperti konser Bruno Mars, menyebabkan lonjakan akses yang sering menyebabkan server lambat hingga crash, serta antrian yang tidak efisien. Untuk mengatasi masalah ini, metode iteratif dan rekursif dalam algoritma sorting dapat diterapkan. Metode iteratif, seperti Bubble Sort menggunakan perulangan sederhana yang lebih hemat memori, tetapi kurang efisien untuk data besar. Di sisi lain, metode rekursif, seperti Merge Sort, membagi data menjadi sub-masalah yang lebih kecil untuk diurutkan dengan efisiensi tinggi (O(n log n)), meskipun membutuhkan memori lebih banyak.


### *Apa algoritma yang akan digunakan?*

Untuk mengatasi masalah efisiensi dalam mengurutkan data pada sistem pembelian tiket konser dengan permintaan tinggi, digunakan dua pendekatan algoritma yang berbeda, yaitu metode iteratif dan rekursif.


*   Algoritma iteratif seperti Bubble Sort  (Iteratif), Algoritma ini mengurutkan data dengan membandingkan dan menukar elemen yang berdekatan hingga seluruh data terurut[1]. Metode iteratif memiliki keunggulan dalam hal penggunaan memori yang lebih efisien, namun memiliki kekurangan dalam hal kecepatan jika diterapkan pada data yang besar[2].
*   Sebaliknya, (Rekursif), Merge Sort membagi data menjadi subarray lebih kecil, mengurutkannya, dan kemudian menggabungkannya[3]. Dan metode rekursif dapat lebih efisien dalam hal pemrosesan, meskipun memerlukan lebih banyak memori untuk menyimpan hasil sementara[2].
*   Secara keseluruhan, pemilihan antara metode iteratif dan rekursif tergantung pada ukuran data dan kebutuhan memori. Metode iteratif lebih sederhana dan hemat memori, tetapi kurang efisien untuk data besar, sementara metode rekursif seperti Merge Sort menawarkan efisiensi lebih tinggi dalam pengurutan data besar, meskipun memerlukan penggunaan memori yang lebih besar.

```python
# Import library
import time

# Fungsi Bubble Sort (iteratif)
def bubble_sort(data):
    n = len(data)
    for i in range(n-1):
        for j in range(n-i-1):
            if data[j][1] > data[j+1][1]:  # Sorting berdasarkan waktu kedatangan
                data[j], data[j+1] = data[j+1], data[j]
    return data

# Fungsi untuk menghitung estimasi waktu pelayanan
def estimate_service(queue, target_name, service_time):
    # Cari posisi dalam antrean
    for idx, (name, arrival_time) in enumerate(queue):
        if name == target_name:
            position = idx + 1
            time_to_be_served = idx * service_time
            return position, time_to_be_served
    return None, None

# Fungsi utama
def main():
    # Data pembeli (nama, waktu kedatangan dalam detik)
    pembeli = [
        ("Andi", 45),
        ("Budi", 30),
        ("Citra", 15),
        ("Dewi", 50),
        ("Eka", 25)
    ]

    # Waktu rata-rata pelayanan per pembeli (dalam detik)
    service_time = 10  # Contoh: 10 detik per pembeli

    # Sorting antrean berdasarkan waktu kedatangan
    sorted_queue = bubble_sort(pembeli[:])

    print("Antrean Saat Ini (Setelah Sorting):")
    for idx, (name, arrival_time) in enumerate(sorted_queue):
        print(f"{idx + 1}. Nama: {name}, Waktu Kedatangan: {arrival_time} detik")

    # Input nama pembeli yang ingin dicek
    target_name = input("\nMasukkan nama Anda untuk mengetahui posisi dalam antrean: ").strip()

    # Estimasi posisi dan waktu pelayanan
    position, time_to_be_served = estimate_service(sorted_queue, target_name, service_time)

    if position is not None:
        print(f"\nHalo {target_name}, posisi Anda dalam antrean adalah urutan ke-{position}.")
        print(f"Anda akan dilayani dalam sekitar {time_to_be_served} detik.")
    else:
        print("\nNama Anda tidak ditemukan dalam antrean.")

# Eksekusi program
if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt

# Fungsi Bubble Sort untuk mengurutkan antrian berdasarkan waktu kedatangan
def bubble_sort(data):
    n = len(data)
    for i in range(n-1):
        for j in range(n-i-1):
            if data[j][1] > data[j+1][1]:  # Sorting berdasarkan waktu kedatangan
                data[j], data[j+1] = data[j+1], data[j]
    return data

# Fungsi untuk menghitung waktu estimasi layanan
def calculate_wait_times(data, waktu_layanan):
    sorted_data = bubble_sort(data)
    wait_times = []
    for i, pembeli in enumerate(sorted_data):
        waktu_hingga_dilayani = waktu_layanan * i  # Menghitung waktu berdasarkan urutan
        wait_times.append((pembeli[0], i + 1, waktu_hingga_dilayani))  # Nama, Posisi, Waktu
    return wait_times

# Fungsi untuk menampilkan grafik
def plot_wait_times(wait_times):
    positions = [item[1] for item in wait_times]
    wait_times_values = [item[2] for item in wait_times]
    names = [item[0] for item in wait_times]

    plt.figure(figsize=(10, 6))
    plt.plot(positions, wait_times_values, marker="o", color="blue", label="Waktu Hingga Dilayani")
    for i, name in enumerate(names):
        plt.text(positions[i], wait_times_values[i], name, fontsize=10, ha='right')
    plt.xlabel("Posisi Antrian")
    plt.ylabel("Waktu Hingga Dilayani (detik)")
    plt.title("Posisi Antrian vs Waktu Hingga Dilayani")
    plt.grid()
    plt.legend()
    plt.show()

# Fungsi utama
def main():
    # Data pembeli (nama, waktu kedatangan dalam detik)
    pembeli = [
        ("Andi", 45),
        ("Budi", 30),
        ("Citra", 15),
        ("Dewi", 50),
        ("Eka", 25)
    ]

    # Waktu layanan rata-rata untuk satu pembeli (detik)
    waktu_layanan = 10

    # Menghitung waktu antrian
    wait_times = calculate_wait_times(pembeli, waktu_layanan)

    # Menampilkan hasil di terminal
    print("Hasil Perhitungan Antrian:")
    for item in wait_times:
        print(f"Nama: {item[0]}, Posisi: {item[1]}, Waktu Hingga Dilayani: {item[2]} detik")

    # Menampilkan grafik
    plot_wait_times(wait_times)

# Eksekusi program
if __name__ == "__main__":
    main()
```

```python
# Import library
import time
import matplotlib.pyplot as plt

# Fungsi Bubble Sort (iteratif)
def bubble_sort(data):
    n = len(data)
    for i in range(n-1):
        for j in range(n-i-1):
            if data[j][1] > data[j+1][1]:  # Sorting berdasarkan waktu kedatangan
                data[j], data[j+1] = data[j+1], data[j]
    return data

# Fungsi Merge Sort (rekursif)
def merge_sort(data):
    if len(data) > 1:
        mid = len(data) // 2
        left_half = data[:mid]
        right_half = data[mid:]

        merge_sort(left_half)
        merge_sort(right_half)

        i = j = k = 0
        while i < len(left_half) and j < len(right_half):
            if left_half[i][1] < right_half[j][1]:
                data[k] = left_half[i]
                i += 1
            else:
                data[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            data[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            data[k] = right_half[j]
            j += 1
            k += 1
    return data

# Fungsi untuk menjalankan sorting dan mengukur waktu
def measure_sort_time(sort_function, data):
    start_time = time.time()
    sort_function(data)
    return time.time() - start_time

# Fungsi utama
def main():
    # Data pembeli (nama, waktu kedatangan dalam detik)
    pembeli = [
        ("Andi", 45),
        ("Budi", 30),
        ("Citra", 15),
        ("Dewi", 50),
        ("Eka", 25)
    ]

    # Variasi ukuran data untuk perbandingan
    data_sizes = [5, 10, 20, 50, 100, 200, 500]
    iterative_times = []
    recursive_times = []

    for size in data_sizes:
        test_data = pembeli * (size // len(pembeli))  # Menggandakan data agar sesuai ukuran
        # Bubble Sort (iteratif)
        iterative_times.append(measure_sort_time(bubble_sort, test_data[:]))
        # Merge Sort (rekursif)
        recursive_times.append(measure_sort_time(merge_sort, test_data[:]))

    # Menampilkan hasil perbandingan waktu eksekusi
    print("Ukuran Data:", data_sizes)
    print("Waktu Iteratif (Bubble Sort):", iterative_times)
    print("Waktu Rekursif (Merge Sort):", recursive_times)

    # Visualisasi menggunakan grafik garis
    plt.figure(figsize=(10, 6))
    plt.plot(data_sizes, iterative_times, label="Bubble Sort (Iteratif)", marker="o", color="blue")
    plt.plot(data_sizes, recursive_times, label="Merge Sort (Rekursif)", marker="s", color="green")
    plt.xlabel("Ukuran Data")
    plt.ylabel("Waktu Eksekusi (detik)")
    plt.title("Perbandingan Waktu Eksekusi Sorting")
    plt.legend()
    plt.grid()
    plt.show()

# Eksekusi program
if __name__ == "__main__":
    main()
```
